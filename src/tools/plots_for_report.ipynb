```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)}, font_scale = 2)
sns.set_style("whitegrid")
save_path = Path("/home/rsa39/rds/rds-mlmi-2020-21-xyBFuSj0hm0/MLMI.2021-22/rsa39/project/RAVQA/data/figures_for_report")
```

```python
# the sample dataframe from the OP
df = pd.DataFrame(
    [
        ['frozen', 'T0-3B (n=5)', 15.78], ['hotpotqa', 'T0-3B (n=5)', 31.07],
        ['frozen', 'T0-3B (n=10)', 20.89], ['hotpotqa', 'T0-3B (n=10)', 34.49],
        # ['frozen', 'T0pp (n=2)', 19.64], ['hotpotqa', 'T0pp (n=2)', 22.5],
        # ['frozen', 'T0pp (n=5)', 22.69], ['hotpotqa', 'T0pp (n=5)', 29.39],
    ],
    columns=['Template', 'VLM', 'VQA Score (%)']
)
df.head()

sns.set_palette("Paired")
zero_shot_barplot = sns.barplot(data=df, x='VLM', y='VQA Score (%)', hue='Template')
zero_shot_barplot.axhline(29.5, ls='--', c="teal", linewidth=4)
sns.despine(left = True)
plt.legend(loc=2, title="Template", bbox_to_anchor=(1.01, 1), borderaxespad=0.)
plt.savefig(save_path / "zero-shot-results.pdf", bbox_inches="tight")

#The plot is shown
plt.show()
```
